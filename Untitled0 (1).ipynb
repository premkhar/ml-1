1  what is a parameter?

     
   In Python, a parameter is a variable defined within the parentheses of a function definition. It acts as a placeholder for a value that will be passed to the function when it is called. Parameters allow functions to accept input and perform operations based on that input. When a function is called, the values passed to it are called arguments, which are then assigned to the corresponding parameters within the function's scope.

15 What is causation? Explain difference between correlation and causation with an example.

ausation means one event directly results in another, while correlation simply indicates a relationship between two events, where one doesn't necessarily cause the other. For example, eating ice cream and getting sunburned are correlated during summer, but eating ice cream doesn't cause sunburns, both are caused by sunny weather.
Here's a more detailed explanation:
Causation:
This implies a direct cause-and-effect relationship where one event (the cause) leads to another event (the effect). For example, if you turn on a light switch, the light turning on is a causal relationship.
Correlation:
This describes a relationship or pattern between two or more variables, meaning they tend to change together, but one doesn't necessarily cause the other. For example, ice cream sales and pool drownings might increase during the summer, but this doesn't mean eating ice cream causes drownings; both are related to the weather.
Key Differences:
Directionality:
Causation implies a direction (A causes B), while correlation doesn't necessarily have a direction.
Strength of Relationship:
Causation is a stronger statement than correlation, as it suggests a direct and influential relationship.
Manipulation:
Causation can be investigated through controlled experiments where one variable is manipulated to see its effect on another. Correlation is often observed in observational data

  17  What is sklearn.linear_model ?

LinearRegression: Ordinary least squares linear regression.
Ridge: Linear least squares with L2 regularization.

ElasticNet: Linear least squares with a combination of L1 and L2 regularization.
LogisticRegression: Logistic regression for binary and multiclass classification.
SGDRegressor: Linear model fitted by minimizing a regularized empirical loss with SGD.
SGDClassifier: Linear classifiers (SVM, logistic regression, a.o.) with SGD training.
